```
import pandas as pd
file_url = 'https://raw.githubusercontent.com/'\
           'PacktWorkshops/The-Data-Science-Workshop/'\
           'master/Chapter10/dataset/ames_iowa_housing.csv'
df1 = pd.read_csv(file_url)
df1.sample(n=5, random_state=5)
```

```
df1.dtypes.unique()
```

```
obj_df =df1.select_dtypes(include='object')
obj_col = obj_df.columns
print(obj_col)
```

```
def describe_obj(df, col_name):
  print(f"\nCOLUMN: {col_name}")
  print(f"{df[col_name].nunique()} different values")
  print(f"List of Values:")
  print(df[col_name].value_counts(dropna=False, normalize=True))
describe_obj(df1, 'MSZoning')
```

```
for col_name in obj_col:
  des = describe_obj(df1, col_name)
  
```

```
num_col = df1.select_dtypes(include= 'number')
num_df = num_col.columns
```

```
def describe_num(df, col_name):
  print(f"\nCOLUMN: {col_name}")
  print(f"Minimum: {df[col_name].min()}")
  print(f"Maximum: {df[col_name].max()}")
  print(f"Average: {df[col_name].mean()}")
  print(f"Median: {df[col_name].median()}")
  print(f"Standard Deviation: {df[col_name].std()}")
describe_num(df1, 'SalePrice')
```

```
for col_name in num_df:
  describe_num(df1, col_name)
```
