```
import pandas as pd
tsv_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter01/Dataset/overall_topten_2012-2013.tsv'
tsv_df=pd.read_csv(tsv_url, skiprows=1, sep='\t')  # tsv is similar to csv, it uses the seperator (\t) as compared to csv that uses (,), sep means separator
tsv_df
```
